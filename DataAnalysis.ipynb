```python
# Check missing values (including special characters)
missing_indicators = ['?', 'N/A', 'missing', 'na', 'null', '', 99999]
missing_values = df.replace(missing_indicators, pd.NA).isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

print("\nMissing values by column:")
for col in df.columns:
    print(f"{col}: {missing_values[col]} ({missing_percentage[col]:.2f}%)")
```
